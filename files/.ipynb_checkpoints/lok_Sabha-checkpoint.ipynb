```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df_2014=pd.read_csv("constituency_wise_results_2014.csv", encoding = 'unicode_escape')
df_2019=pd.read_csv("constituency_wise_results_2019.csv", encoding = 'unicode_escape')
```

```python
pd.isnull(df_2019).sum()
```

```python
df_2014.drop(['Unnamed: 13','Unnamed: 14','Unnamed: 15','party_symbol'], axis=1, inplace= True,  errors='ignore')
df_2019.drop(['party_symbol'], axis=1, inplace= True,  errors='ignore')
```

```python

df_2014.fillna(" ",inplace = True)
df_2019.fillna(" ",inplace = True)
```

```python
# Combine the datasets to create a single list of unique constituencies
df_constituencies = pd.concat([df_2014[['state', 'pc_name']], df_2019[['state', 'pc_name']]]).drop_duplicates().reset_index(drop=True)
df_constituencies['constituency_id'] = df_constituencies.index + 1  # Create a primary key

# Merge constituency IDs into the original datasets
df_2014 = df_2014.merge(df_constituencies, on=['state', 'pc_name'], how='left')
df_2019 = df_2019.merge(df_constituencies, on=['state', 'pc_name'], how='left')
```

# List top 5 constituencies for two major national parties where they have gained voteshare in 2019 as compared to 2014.

FOR BJP

```python
bjp14=df_2014[df_2014['party']=='BJP']
bjp19=df_2019[df_2019['party']=='BJP']

bjp_2019=bjp19.groupby('pc_name')['total_votes'].sum().reset_index()

bjp_2014=bjp14.groupby('pc_name')['total_votes'].sum().reset_index()

bjp_2019.rename(columns={'total_votes': 'total_votes_2019'}, inplace=True)
bjp_2014.rename(columns={'total_votes': 'total_votes_2014'}, inplace=True)
df_BJP = pd.merge(bjp_2019, bjp_2014, how='left',on='pc_name')
df_BJP.fillna(0, inplace=True)
df_BJP['vote_change'] = df_BJP['total_votes_2019'] - df_BJP['total_votes_2014']
df_bjp=df_BJP.sort_values(by='vote_change', ascending=False)
```

```python
df_bjp.head(5)
```

```python
top_vote=df_bjp.head(5)
top_vote
import seaborn as sns

ax=sns.barplot(x=top_vote['vote_change'],y=top_vote['pc_name'], data=top_vote)
sns.set(rc={'figure.figsize':(12, 8)}, style="whitegrid")

ax.set_title('Vote Change for BJP in 2019 vs 2014', fontsize=16)
ax.set_xlabel('Vote Change', fontsize=14)
ax.set_ylabel('Parliamentary Constituency', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
```

FOR INC

```python
INC14=df_2014[df_2014['party']=='INC']
INC19=df_2019[df_2019['party']=='INC']

INC_2019=INC19.groupby('pc_name')['total_votes'].sum().reset_index()

INC_2014=INC14.groupby('pc_name')['total_votes'].sum().reset_index()

INC_2019.rename(columns={'total_votes': 'total_votes_2019'}, inplace=True)
INC_2014.rename(columns={'total_votes': 'total_votes_2014'}, inplace=True)
df_INC = pd.merge(INC_2019, INC_2014, on='pc_name',how ='left')
df_INC.fillna(0, inplace=True)
df_INC['vote_change'] = df_INC['total_votes_2019'] - df_INC['total_votes_2014']
df_inc=df_INC.sort_values(by='vote_change', ascending= False)

```

```python
df_inc.head(5)
```

```python
top_vote=df_inc.head(5)
import seaborn as sns

ax=sns.barplot(x=top_vote['vote_change'],y=top_vote['pc_name'], data=top_vote)
sns.set(rc={'figure.figsize':(12, 8)}, style="whitegrid")
ax.set_title('Vote Change for INC in 2019 vs 2014', fontsize=16)
ax.set_xlabel('Vote Change', fontsize=14)
ax.set_ylabel('Parliamentary Constituency', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Is there a correlation between postal votes % and voter turnout %?

```python
df=df_2019.groupby('pc_name').agg({'total_votes':'sum','postal_votes': 'sum'}).reset_index()
dff = df_2019.groupby('pc_name')['total_electors'].unique().reset_index()
r = pd.merge(df, dff, on='pc_name')

r['vtr'] = (r['total_votes'] / r['total_electors']) * 100
```

```python
corr = pd.read_csv('f.csv')
```

```python
corr
```

```python
corr.drop(['Unnamed: 0'], axis=1, inplace= True,  errors='ignore')
```

```python
import numpy as np
import pandas as pd
corr['postal_votes'] = pd.to_numeric(corr['postal_votes'])
corr['vtr'] = pd.to_numeric(corr['vtr'])
coff =  np.corrcoef(corr['postal_votes'],corr['vtr'])
```
